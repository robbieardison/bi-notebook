```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('./downloads/netflix data/titles.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df2 = df.query("type=='MOVIE'").reset_index()
df2
```

```python
df3 = df2.groupby(['production_countries']).mean().sort_values(by='imdb_score', ascending=False).iloc[:10].reset_index()
df3
```

```python
x = df3['production_countries'].str[1:-1]
y = df3['imdb_score']

```

```python
fig = plt.figure(figsize=(50,20))
plt.barh(x, y, color='green')
plt.xlabel('Average IMDB Score', fontsize=30)
plt.ylabel('Country', fontsize=30)
plt.xticks(fontsize=40)
plt.yticks(fontsize=40)
plt.title('Top 10 Highest Rated IMDB Movies by Production Countries', fontsize=70)
```
