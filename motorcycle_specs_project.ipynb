```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('./downloads/archive/all_bikez_curated.csv').dropna().reset_index()
df = pd.DataFrame(data)
df.head()
```

```python
df2 = df.query("Category == 'Naked bike'")
df3 = df2.groupby('Brand').mean().sort_values(by='Power (hp)', ascending=False).iloc[:10].reset_index()
df3
```

```python
x=df3['Brand']
y=df3['Power (hp)']
fig = plt.figure(figsize=(30,20))
plt.bar(x, y, color='green')
plt.title(label='T0P 10 NAKED BIKES BY AVERAGE HORSEPOWER')
plt.xlabel('Brands', fontsize=40)
plt.ylabel('Average Horsepower', fontsize=40)
plt.show()
```
